```python
#read csv files
DataFileAddress = "/content/drive/MyDrive/Colab Notebooks/Analysis/Analysis_all_data_retrained_for_processing.csv"
#Ethnicity data
Ethnicity_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Analysis/Ethnicity_data.csv')
df = pd.read_csv(DataFileAddress)  #main data csv files
```

```python
#merge 3 columns (Code 1, Code 2, and Code 3) data to one data column
new_rows = []
output = pd.DataFrame()
for index, row in df.iterrows():
    if not pd.isnull(row[7]):
        new_row = row.copy()
        new_row[6] = row[7]
        new_row[7] = None
        new_rows.append(new_row)
    if not pd.isnull(row[8]):
        new_row = row.copy()
        new_row[6] = row[8]
        new_row[8] = None
        new_rows.append(new_row)

# append to the main file
output = df.append(new_rows)
```

```python
# add ethnicity data

RowData = pd.DataFrame()

for x in output.Filename:
    RowData = RowData.append(Ethnicity_data.loc[Ethnicity_data.FileNames==x])
```

```python
# reset indeces to save in a csv file
output = output.reset_index(drop=True)
RowData = RowData.reset_index(drop=True)

result = pd.concat([output, RowData], axis=1)
result['Code1'] = result['Code1'].str.strip()  # remove extrace spaces
result = result[result['Code1'] != ''] #Remove rows with empty strings in column 
result=result.dropna(subset=['Code1'])
```
